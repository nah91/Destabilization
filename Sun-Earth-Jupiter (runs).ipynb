```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy
from scipy.integrate import odeint

### Define constants:
G = 6.67408 * 10**(-11)      # m**3 / (kg * s**2)     [Gravitational constant]
T_Earth = 3.154 * 10**7      # seconds                [number of seconds per Earth year]
T_Jupiter = 3.74017 * 10**8  # seconds                [number of seconds per Jupiter year]

M0 = 1.989 * 10**30          # kg                     [Mass of Sun]
R3 = 695500000               # m                      [Radius of Sun]

M3 = 5.972 * 10**24          # kg                     [Mass of Earth]
R3 = 6378100                 # m                      [Radius of Earth]
D3 = 1.496 * 10**11          # m                      [Orbital distance Earth-Sun]
V3 = np.sqrt(G*M0/D3)        # m                      [Orbital velocity of Earth]

M5 = 1.898 * 10**27          # kg                     [Mass of Jupiter]
R5 = 69911000                # m                      [Radius of Jupiter]
D5 = 7.786 * 10**11          # m                      [Orbital distance Jupiter-Sun]
V5 = np.sqrt(G*M0/D5)        # m                      [Orbital velocity of Jupiter]

rho = 2000                   # kg / m**3              [Density of Asteroid]
R = 637810                   # m                      [Radius of Asteroid]
Vol = (4/3)*np.pi*(R**3)     # m**3                   [Volume of Asteroid]
M = rho * Vol                # kg                     [Mass of Asteroid]
D = 1.496 * 10**11           # m                      [Orbital distance Asteroid-Sun]
V = np.sqrt(G*M0/D)          # m                      [Orbital velocity of Asteroid]
print('mass of asteroid: {}'.format(M))
```

```python
def genOrbitalMotion(y_arr, t):
    x0, vx0, y0, vy0, x3, vx3, y3, vy3, x5, vx5, y5, vy5 = y_arr
    x0_3, x3_0, y0_3, y3_0 = x0 - x3, x3 - x0, y0 - y3, y3 - y0
    x0_5, x5_0, y0_5, y5_0 = x0 - x5, x5 - x0, y0 - y5, y5 - y0
    x3_5, x5_3, y3_5, y5_3 = x3 - x5, x5 - x3, y3 - y5, y5 - y3
    r0_3 = np.sqrt(x0_3**2 + y0_3**2)
    r0_5 = np.sqrt(x0_5**2 + y0_5**2)
    r3_5 = np.sqrt(x3_5**2 + y3_5**2)
    r3_0, r5_0, r5_3 = r0_3, r0_5, r3_5
    return np.array([vx0, -G * (((M3*x0_3) / r3_0**3) + ((M5*x0_5) / r5_0**3)),
                     vy0, -G * (((M3*y0_3) / r3_0**3) + ((M5*y0_5) / r5_0**3)),
                     vx3, -G * (((M0*x3_0) / r0_3**3) + ((M5*x3_5) / r5_3**3)),
                     vy3, -G * (((M0*y3_0) / r0_3**3) + ((M5*y3_5) / r5_3**3)),
                     vx5, -G * (((M0*x5_0) / r0_5**3) + ((M3*x5_3) / r3_5**3)),
                     vy5, -G * (((M0*y5_0) / r0_5**3) + ((M3*y5_3) / r3_5**3))])

def orbitalMotion(y_arr, t):
    ans_arr = odeint(genOrbitalMotion, y_arr, t)
    return np.array(ans_arr)
```

```python
N = 100000
t3 = np.linspace(0, T_Earth, N)
t5 = np.linspace(0, T_Jupiter, N)

y_arr = np.array([                       0.0,                         0.0,                     0.0,                        0.0,
                  D3*np.cos(np.pi*(120/180)), -V3*np.sin(np.pi*(120/180)), D3*np.sin(np.pi*(120/180)), V3*np.cos(np.pi*(120/180)), 
                  D5*np.cos(np.pi*(240/180)), -V5*np.sin(np.pi*(240/180)), D5*np.sin(np.pi*(240/180)), V5*np.cos(np.pi*(240/180))])
```

```python
Period_Earth = orbitalMotion(y_arr, t3)
Period_Jupiter = orbitalMotion(y_arr, t5)
```

```python
plt.plot(Period_Earth[:,0], Period_Earth[:,2],  'ro', color='y', label='Sun')
plt.plot(Period_Earth[:,4], Period_Earth[:,6],  'ro', color='b', label='Earth')
plt.plot(Period_Earth[:,8], Period_Earth[:,10], 'ro', color='r', label='Jupiter')
plt.title('Orbits of Sun-Earth-Jupiter, 1 Earth Period')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.axes().set_aspect('equal', 'datalim')
plt.show()

plt.plot(Period_Jupiter[:,0], Period_Jupiter[:,2],  'ro', color='y', label='Sun')
plt.plot(Period_Jupiter[:,4], Period_Jupiter[:,6],  'ro', color='b', label='Earth')
plt.plot(Period_Jupiter[:,8], Period_Jupiter[:,10], 'ro', color='r', label='Jupiter')
plt.title('Orbits of Sun-Earth-Jupiter, 1 Jupiter Period')
plt.xlabel('x [m]')
plt.ylabel('y [m]')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.axes().set_aspect('equal', 'datalim')
plt.show()
```
